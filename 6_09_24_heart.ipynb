```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;

from google. colab import drive
drive.mount('/content/drive')



```

```python
df=pd.read_csv("/content/drive/MyDrive/heart.csv" )
df.head()
```

```python
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
sns.countplot(x=df.target,data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target Distributions");
```

```python
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target Distributions");
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend(["haven't disease","have disease"])
plt.title('heart disease frequency for ages')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
#ans.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.displot(df.age[df['target']==0],bins=30,color='#1CA53B',label='Haven\'t Disease')
sns.displot(df.age[df['target']==1],bins=30,color='red',label='Have Disease')
plt.legend()
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='Black')
sns.distplot(df.age[df['target']==1],bins=30,color='pink')
plt.legend()
plt.title("Heart disease distributin for ages")
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.show()
```
